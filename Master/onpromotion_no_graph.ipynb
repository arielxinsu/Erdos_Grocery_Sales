```python
import pandas as pd
```

```python
#Here combined_v2.csv is the second version of the data combined by Maksim
SI=pd.read_csv('combined_v2.csv')
SI_new=pd.read_csv('merged_train.csv')
og=pd.read_csv('Data/train.csv')
```

```python
# Drop rows with NaN values in the 'family' column
SI = SI.dropna(subset=['family'])

# Display the resulting DataFrame
#SI.head()
```

```python
# Calculate the correlation between 'sales' and 'onpromotion'
correlation_sales_onpromotion = SI['sales'].corr(SI['onpromotion'])

# Display the result
print(correlation_sales_onpromotion.round(2))
```

```python
# Count entries with non-zero 'onpromotion' for each 'family'
non_zero_onpromotion_counts = SI[SI['onpromotion'] > 0].groupby('family').size().reset_index(name='non_zero_onpromotion_count')

# Display the result
non_zero_onpromotion_counts.sort_values(by='non_zero_onpromotion_count')
```

```python
# Filter rows where 'family' is 'BABY CARE' and 'onpromotion' is greater than 0
baby_care_dates = SI[(SI['family'] == 'BABY CARE') & (SI['onpromotion'] > 0)]['date']

# Display the unique dates
baby_care_dates = baby_care_dates.unique()
len(baby_care_dates)
```

```python
# Filter rows where 'family' is 'HOME APPLIANCES' and 'onpromotion' is greater than 0
home_appliances_dates = SI[(SI['family'] == 'HOME APPLIANCES') & (SI['onpromotion'] > 0)]['date']

# Display the unique dates
home_appliances_dates = home_appliances_dates.unique()

# Print the result
print("Dates where 'HOME APPLIANCES' items were on promotion:")
print(len(home_appliances_dates))
```

```python
# Group by 'family' and find the maximum of 'onpromotion' for each group
max_onpromotion_per_family = SI.groupby('family')['onpromotion'].max().reset_index()

# Display the result
#max_onpromotion_per_family
max_onpromotion_per_family.sort_values(by='onpromotion')
```

```python
# Calculate the correlation between 'sales' and 'onpromotion' for each 'family'
correlation_per_family = SI.groupby('family').apply(lambda x:round(x['sales'].corr(x['onpromotion']), 2)).reset_index(name='correlation')

# Display the result
#correlation_per_family.sort_values(by='correlation',ascending=True)
```

```python
# Count entries with non-zero 'onpromotion' for each 'family'
non_zero_onpromotion_counts = SI[SI['onpromotion'] > 0].groupby('family').size().reset_index(name='non_zero_onpromotion_count')

# Sort the result by the non-zero 'onpromotion' count in descending order
non_zero_onpromotion_counts = non_zero_onpromotion_counts.sort_values(by='non_zero_onpromotion_count', ascending=True).reset_index(drop=True)

# Display the result
#non_zero_onpromotion_counts
```

```python
# Calculate the correlation between 'sales' and 'onpromotion' for each 'day_of_the_week'
correlation_per_day = SI.groupby('day_of_the_week').apply(lambda x: x['sales'].corr(x['onpromotion'])).reset_index(name='correlation')

# Display the result
correlation_per_day.sort_values(by='correlation')
```

```python
# Calculate the correlation between 'sales' and 'onpromotion' for each 'store_nbr'
correlation_per_store = SI.groupby('store_nbr').apply(lambda x: round(x['sales'].corr(x['onpromotion']),2)).reset_index(name='correlation')

# Display the result
#correlation_per_store.sort_values(by='correlation')
```

```python
# Calculate the average sales for entries with and without promotions
average_sales_with_promotion = SI[SI['onpromotion'] > 0]['sales'].mean()
average_sales_without_promotion = SI[SI['onpromotion'] == 0]['sales'].mean()

# Calculate the percentage increase in sales due to promotions
percentage_increase = ((average_sales_with_promotion - average_sales_without_promotion) / average_sales_without_promotion) * 100

# Display the results with explanations
print(f"Average sales with promotion: {average_sales_with_promotion:.2f}")
print(f"Average sales without promotion: {average_sales_without_promotion:.2f}")
print(f"Percentage increase in sales due to promotions: {percentage_increase:.2f}%")
```

```python
# Calculate average units sold with and without promotions for each 'family' and the percentage increase
promotion_impact_per_family = SI.groupby('family').apply(lambda x: pd.Series({
    'avg_units_with_promotion': x[x['onpromotion'] > 0]['sales'].mean(),
    'avg_units_without_promotion': x[x['onpromotion'] == 0]['sales'].mean()
})).reset_index()

# Calculate the percentage increase in units sold due to promotions for each family
promotion_impact_per_family['percentage_increase'] = (
    (promotion_impact_per_family['avg_units_with_promotion'] - promotion_impact_per_family['avg_units_without_promotion']) / 
    promotion_impact_per_family['avg_units_without_promotion'] * 100
)

# Format the display for readability
promotion_impact_per_family['avg_units_with_promotion'] = promotion_impact_per_family['avg_units_with_promotion'].apply(lambda x: f"{x:,.2f} units")
promotion_impact_per_family['avg_units_without_promotion'] = promotion_impact_per_family['avg_units_without_promotion'].apply(lambda x: f"{x:,.2f} units")
promotion_impact_per_family['percentage_increase'] = promotion_impact_per_family['percentage_increase'].apply(lambda x: f"{x:.2f}%")


# Remove the percentage symbol and convert to numeric
promotion_impact_per_family['percentage_increase'] = promotion_impact_per_family['percentage_increase'].str.rstrip('%').astype(float)

# Sort by 'percentage_increase' in descending order
promotion_impact_per_family = promotion_impact_per_family.sort_values(by='percentage_increase', ascending=False)

# Display the result
#promotion_impact_per_family.sort_values(by='percentage_increase')
```
